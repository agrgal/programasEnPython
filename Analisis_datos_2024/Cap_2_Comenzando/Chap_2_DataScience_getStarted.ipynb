```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pt
```

```python
data = {'año': [2010 , 2011 , 2012 ,2010 , 2011 , 2012 ,2010 , 2011 , 2012]
         ,'equipo': ['FCBarcelona', 'FCBarcelona','FCBarcelona', 'RMadrid'
                   ,'RMadrid', 'RMadrid','ValenciaCF', 'ValenciaCF','ValenciaCF']
         ,'victorias':[30 , 28 , 32 , 29 , 32 , 26 , 21 , 17 , 19]
         ,'empates':[6 , 7 , 4 , 5 , 4 , 7 , 8 , 10 , 8]
         ,'derrotas': [2 , 3 , 2 , 4 , 2 , 5 , 9 , 11 , 11]
       }
football=pd.DataFrame( data , columns = ['año','equipo','victorias','empates','derrotas'])
```

```python
football
```

```python
# @title equipo vs empates

from matplotlib import pyplot as plt
import seaborn as sns
figsize = (12, 1.2 * len(football['equipo'].unique()))
plt.figure(figsize=figsize)
sns.violinplot(football, x='empates', y='equipo', inner='stick', palette='Dark2')
sns.despine(top=True, right=True, bottom=True, left=True)
```
